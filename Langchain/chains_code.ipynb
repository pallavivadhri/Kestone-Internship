```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
GROQ_API_KEY= " "
model= ChatGroq(model="llama-3.3-70b-versatile",api_key=GROQ_API_KEY)
```

```python

from langchain_core.prompts import PromptTemplate

movie_prompt_template = PromptTemplate.from_template ("tell me a/an {language} movie title of the {genre} genre")
prompt = movie_prompt_template.invoke ({"language":"English" , "genre" :"horror"})
response = model.invoke(prompt)
print(response.content)
```

```python
#SEQUENTIAL/EXTENDED CHAINING
# | --> is pipe operator
from langchain_core.output_parsers import StrOutputParser 
movie_template_chain = movie_prompt_template | model
chain_response = movie_template_chain.invoke({"language":"English" , "genre" :"horror"})
print(chain_response)
```

```python
movie_template_chain = movie_prompt_template | model | StrOutputParser()
chain_response = movie_template_chain.invoke({"language":"English" , "genre" :"horror"})
print(chain_response)
```

```python
movie_summary_prompt = PromptTemplate.from_template("give me the summary of the movie {movie_title} in 4 points")

composed_chain = {"movie_title": movie_template_chain} | movie_summary_prompt | model |StrOutputParser()

response_1 = composed_chain.invoke( {"language":"English" , "genre" :"horror"} )

print(response_1)
```

```python
from langchain_core.runnables import RunnableLambda
print_title_step = RunnableLambda ( lambda x :(print(x), x)[1])
#print_title_step = RunnableLambda ( lambda x : print(x["movie_title"]))

composed_chain = {"movie_title": movie_template_chain} | print_title_step | movie_summary_prompt | model |StrOutputParser()
response_1 = composed_chain.invoke( {"language":"English" , "genre" :"horror"} )

print(response_1)
```

```python
#using Runnable sequence instead of pipe operator
from langchain_core.runnables import RunnableSequence
movie_summary_prompt = PromptTemplate.from_template("give me the summary of the movie {movie_title} in 4 points")

composed_chain = RunnableSequence({"movie_title": movie_template_chain} , movie_summary_prompt , model ,StrOutputParser())

response_1 = composed_chain.invoke( {"language":"English" , "genre" :"horror"} )

print(response_1)
```
